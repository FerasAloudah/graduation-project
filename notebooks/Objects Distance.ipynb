```python
import pandas as pd
```

```python
df = pd.read_csv('../metadata/relationships/relationships_none_2020-04-10_15-32-55.csv', nrows=700_000)
```

```python
df.head()
```

```python
df['TopLeft'] = ((df['XMin2'] - df['XMin1'])**2 + (df['YMin2'] - df['YMax1'])**2)**0.5
df['TopRight'] = ((df['XMax2'] - df['XMax1'])**2 + (df['YMax2'] - df['YMax1'])**2)**0.5
df['BottomLeft'] = ((df['XMin2'] - df['XMin1'])**2 + (df['YMin2'] - df['YMin1'])**2)**0.5
df['BottomRight'] = ((df['XMax2'] - df['XMax1'])**2 + (df['YMin2'] - df['YMin1'])**2)**0.5
df['Center'] = (df['TopLeft'] + df['TopRight'] + df['BottomLeft'] + df['BottomRight']) / 4
```
