```python
import pandas as pd
import glob
```

### Appending all the data togther

```python
v = pd.read_csv('../relationships/oidv6-validation-annotations-vrd.csv')
te = pd.read_csv('../relationships/oidv6-test-annotations-vrd.csv')
# t = pd.read_csv('../relationships/oidv6-train-annotations-vrd.csv')


df1 = t[t['RelationshipLabel'] != 'is']
df2 = v[v['RelationshipLabel'] != 'is']
df3 = te[te['RelationshipLabel'] != 'is']

df = df1.append(df2).append(df3).sort_values(by='LabelName1')
df.head()
```

```python
df.RelationshipLabel.value_counts()
```

### Changing all 'under' to 'on'

```python
df.loc[df.RelationshipLabel == 'under','RelationshipLabel'] = 'on'
```

```python
df.RelationshipLabel.value_counts()
```

```python
names = []
for name, group in df.groupby(['LabelName1', 'LabelName2']):
    names.append(name)

len(names)
```

### Counting relationships:  RelationshipCount

```python
r = df[['LabelName1', 'LabelName2', 'RelationshipLabel']].pivot_table(index=['LabelName1', 'LabelName2', 'RelationshipLabel'], aggfunc='size')
r = r.to_frame().reset_index()
r.columns = ['LabelName1', 'LabelName2', 'RelationshipLabel', 'RelationshipCount']
r.head(15)
```

### Counting labels: LabelCount

```python
l = df[['LabelName1', 'LabelName2']].pivot_table(index=['LabelName1', 'LabelName2'], aggfunc='size')
l = l.to_frame().reset_index()
l.columns = ['LabelName1', 'LabelName2', 'LabelCount']
l.head()
```

```python
count = r.merge(l)
count.head(20)
```
