```python
import pandas as pd
import glob
```

### Loading data and appending them data in  a df

```python
t = pd.read_csv('../metadata/relationships/oidv6-train-annotations-vrd.csv')
df1 = t[t['RelationshipLabel'] != 'is']
v = pd.read_csv('../metadata/relationships/oidv6-validation-annotations-vrd.csv')
df2 = v[v['RelationshipLabel'] != 'is']
te = pd.read_csv('../metadata/relationships/oidv6-test-annotations-vrd.csv')
df3 = te[te['RelationshipLabel'] != 'is']


df = df1.append(df2).append(df3)
df.head()
```

### Calculating obj1 and obj2 percetage

```python
df['obj1'] = ((df['XMax1'] - df['XMin1']) * (df['YMax1'] - df['YMin1']))
df['obj2'] = ((df['XMax2'] - df['XMin2']) * (df['YMax2'] - df['YMin2']))
```
